## 1. Cuadrante en el plano carteciano (1.0)

image.png 

Dados dos números reales distintos de cero **x** y **y** que representen las coordenadas de un punto P(x, y) ∈ ℝ^2. Determinar el proceso para identificar a que cuadrante del plano cartesiano pertenece el punto P. 

````
def cuadrante(x, y):
  cuadrante=""
  if(x>0 and y>0):
    cuadrante="Cuadrante I"
  elif(x>0 and y<0):
    cuadrante="Cuadrante IV"
  elif(x<0 and y>0):
    cuadrante="Cuadrante II"
  elif(x<0 and y<0):
    cuadrante="Cuadrante III"
  else:
    cuadrante="Número 0"
  return cuadrante
````

````
cuadrante(-5,-50)
````

## 2. Valor absoluto (0.5)
Defina el procedimiento que permite calcular el valor absoluto de un número real **x**. La función definida debe retornar la distancia desde el valor x hasta el origen del plano. 

````
def valor_absoluto(x):
  if(x>=0):
    return int(x)
  elif(x<0):
    return (x*-1)
````

## 7. Busqueda en lista

Usted entra a trabajar en una cadena de supermercados y para impresionar a su jefe usted decide crear un proyecto aplicando sus conocimientos en programación de computadores. El programa que usted quiere diseñar debe cumplir el objetivo de buscar un producto en la lista del inventario del supermercado. Cuando encuentre el producto detener la busqueda y retornar el siguiente mensaje: "*Item encontrado*". En el caso que no se encuentre el producto se debe imprimir el mensaje: “*Item NO encontrado en la lista*”.

````
def busqueda_inventario(producto):
  inventario = ['arroz', 'papa', 'pan', 'pasta', 'cereal', 'harina', 'huevos', 
                'queso', 'jamón', 'pollo', 'carne', 'pescado', 'mariscos',
                'manzana', 'pera','piña', 'granadilla', 'durazno', 'fresa',
                'tomate', 'lechuga', 'cebolla', 'aguacate', 'champiñones'
                'cepillo de dientes', 'crema dental', 'shampoo', 'jabón']
  for i in inventario:
    if(i==producto):
      return "Item encontrado"
  else:
    return "Item NO encontrado en la lista"
  
````

````
busqueda_inventario('Crema dental')
````
