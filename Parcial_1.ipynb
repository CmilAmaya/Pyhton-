1. Escribir un programa que calcule el índice de masa corporal (IMC). La función definida debe recibir como argumento el peso (en kg) y estatura (en metros) de una persona. Recuerde que el IMC está definido según la siguiente ecuación: `IMC = peso[kg]/(estatura[m]^2)`

````
def imc(peso, estatura):
  '''
  Entrada:
  peso (float): Peso del usuario
  estatura (float): Estatura del usuario

  Salida:
  (float): Índice de masa corporal
  '''
  imc=peso/(estatura**2)
  return imc
````

````
imc(32,1.65)
````

4. Escribir una función que calcule el mínimo común múltiplo entre dos números que ingresan como parámetro.

````
def mcm(n, m):
  '''
  Entrada:
  n (float): Primer número a ser evaluado
  m (float): Segundo número a ser evaluado

  Salida:
  (float): Mínimo común múltiplo entre n y m
  '''
  z= max(n, m)
  while True:
    if(z%n==0) and (z%m==0):
      return z
    z+=1
  
````

````
mcm(5,20)
````

5. Escribir un programa que recibe como parámetro un número entero positivo (va a ser mayor de 2) y verifica si el un número es primo o no. El resultado de verifica en el contexto booleano.



````
def numero_primo(n):
  '''
  Entrada:
  n (float): Número a ser evaluado

  Salida:
  (bool): Verificacion de si n es un número primo
  '''
  for i in range(2, n):
        if n % i == 0:
            return False
  return True
    
````

````
numero_primo(6)
````

6. Escribir una función que verifique si un número es perfecto o no. Un número perfecto es un entero positivo que es igual a la suma de sus divisores positivos (excluyendo el mismo número). Equivalentemente, un número perfecto es un número que es la mitad de la suma de todos sus divisores positivos (incluyendo el mismo). **Ejemplo:** E primer número perfecto es 6, porque ` 1 + 2 + 3 = ( 1 + 2 + 3 + 6 ) / 2 = 6 `

````
def numero_perfecto(n):
  '''
  Entrada:
  n (int): Número entero positivo a ser evaluado

  Salida:
  (bool): Verificación de si el número es perfecto
  '''
  suma=0
  for i in range(1, n+1):
        if (n%i==0):
          suma=suma+i 
  s=suma/2
  if(s==n):
    print(True)
  else:
    print(False) 
  return 
````

````
numero_perfecto(2)
````

7. Establezca una función que permita calcular el volumen del sólido que se presenta a continuación (a partir de los radios y la altura): Imagen2.png

````
import math
def volumen(r1, r2, h, pi):
  '''
  Entrada:
  r1: Radio de la esfera
  r2: Radio del cono
  h: Altura del cono
  pi: Valor de la constante matemática pi

  Salida:
  (float): Volumen total del sólido completo
  '''
  esfera=(4/3)*(pi*r1**3)
  cono=(pi*r2**2*h)/3
  v_total=esfera+cono
  return v_total
````

8. Diseñar una función que calcule el valor final a pagar por un prestamo de P cantidad de pesos, el cual tiene un plazo de seis meses, con un interés compuesto de 3% mensual. Defina los valores como parametros por defecto para poder calcular el valor final a pagar con diferentes plazos y tasa de interés. Recuerde la formula de interes compuesto es: `F = P(1+i)^n `



````
def interes_compuesto(P, n = 6, i = 0.03):
  '''
  Entrada:
  P: Valor del prestamo (Capital inicial)
  n: Plazo de pago del prestamo en meses
  i: tasa de interés mensual

  Salida:
  (float):Valor final a pagar incluyendo intereses (Capital final)
  '''
  F = P*(1+i)**n
  return F
````

````
interes_compuesto(1000000)
````

10. Los teléfonos de una empresa tienen el siguiente formato `prefijo-número-extension`. El prefijo es el código del país, que consiste en el signo `+` seguido de los dígitos. Y la extensión  tiene dos dígitos ubicados al final del npumero telefónico. Usted debe escribir un programa que muestre por pantalla el número de teléfono sin el prefijo y la extensión.

**Ejemplo:**

```
Input: +34-913724710-56
Output: 913724710
```



````
def num_telefonico(tel):
  '''
  Entrada:
  tel (str): Número telefónico con el fomrato de la empresa

  Salida:
  (str): Npumero telefónico sin prefijo ni extensión
  '''
  tel=tel.strip("+123456780")
  tel=tel.strip("-")
  return tel
````
